```python
import re

def extract_screen_size_inch(value):
    # Buscar cualquier secuencia de números, incluyendo decimales, seguida de "inch" o "inches"
    match = re.search(r'(\d*\.?\d+)\s*(?:inch|inches)', value, flags=re.IGNORECASE)
    if match:
        return float(match.group(1))
    else:
        return None

# Aplicar la función a la columna "Screen_Size" para extraer los valores en pulgadas
df3['Screen_Size_inch'] = df3['Screen_Size'].apply(extract_screen_size_inch)

df3
```

```python
# Eliminar filas con valores no válidos en la columna 'Storage'
df5 = df5.dropna(subset=['Storage'])

# Convertir los valores de almacenamiento de terabytes a gigabytes
df5['Storage_GB'] = df5['Storage'].apply(lambda x: x * 1000 if x > 1 else x)

# Calcular el tamaño de almacenamiento promedio en gigabytes
storage_avg_GB = df5['Storage_GB'].mean()

print("El tamaño de almacenamiento promedio de las laptops en el conjunto de datos es de aproximadamente", storage_avg_GB, "GB.")
```

```python
# Obtener el recuento de cada valor en la columna 'Touch_Screen'
touch_screen_counts = df5['Touch_Screen'].value_counts()

# Calcular el promedio de cada categoría
touch_screen_avg_no = touch_screen_counts['No'] / len(df5) * 100
touch_screen_avg_yes = touch_screen_counts['Yes'] / len(df5) * 100

print("El promedio de laptops sin pantalla táctil es aproximadamente:", touch_screen_avg_no)
print("El promedio de laptops con pantalla táctil es aproximadamente:", touch_screen_avg_yes)
```
